```
from jamo import h2j, j2hcj 
def get_jongsung_TF(sample_text):
    sample_text_list = list(sample_text) 
    last_word = sample_text_list[-1] 
    last_word_jamo_list = list(j2hcj(h2j(last_word))) 
    last_jamo = last_word_jamo_list[-1] 

    jongsung_TF = "T" 

    if last_jamo in ['ㅏ', 'ㅑ', 'ㅓ', 'ㅕ', 'ㅗ', 'ㅛ', 'ㅜ', 'ㅠ', 'ㅡ', 'ㅣ', 'ㅘ', 'ㅚ', 'ㅙ', 'ㅝ', 'ㅞ', 'ㅢ', 'ㅐ,ㅔ', 'ㅟ', 'ㅖ', 'ㅒ']:
        jongsung_TF = "F" 
    return jongsung_TF
```

```
from konlpy.tag import Mecab
from gensim.models import Word2Vec
from google.colab import drive
drive.mount('/content/gdrive')

model = Word2Vec.load('/content/word2vec')

mecab = Mecab()


furniture_category = ['패브릭소파', '천연/인조가죽소파', '모듈식소파', '풋스툴/쿠션형스툴', '패브릭암체어', '패브릭카우치', 
                      '천연가죽암체어', '천연가죽카우치', '인조가죽암체어', '리클라이너', '라탄암체어', '어린이암체어', '가죽암체어',
                      '더블/퀸/킹침대', '싱글/수퍼싱글+침대', '유아용침대', '간이침대/데이베드', '로프트/2층침대',
                      '소파베드', '디반침대', '여닫이옷장', '어린이옷장', '오픈형옷장', '커피테이블/보조테이블', 
                      '바테이블', '카페테이블', '식탁', '책상/컴퓨터책상', '침대협탁', '콘솔테이블', '화장대', '어린이테이블', 
                      '기저귀교환대', '식탁의자', '바의자', '카페의자','사무용의자', '스툴/벤치', '영아용의자', '스텝스툴/사다리', 
                      '어린이의자', '서랍장', '어린이서랍장', '선반유닛', '수납장', '장식장', '선반/스탠드',
                      '수납가구', '커튼', '블라인드', '샤워커튼', '패널커튼']

sofa = ['패브릭소파', '천연/인조가죽소파', '모듈식소파', '풋스툴/쿠션형스툴', '소파베드', '리클라이너']
bed = ['더블/퀸/킹침대', '싱글/수퍼싱글+침대', '유아용침대', '간이침대/데이베드', '로프트/2층침대', '소파베드', '디반침대']
closet = ['여닫이옷장', '어린이옷장', '오픈형옷장']
table = ['커피테이블/보조테이블', '바테이블', '카페테이블', '식탁', '책상/컴퓨터책상', '침대협탁', '콘솔테이블', '화장대', '어린이테이블', '기저귀교환대']
chair = ['식탁의자', '바의자', '카페의자','사무용의자', '스툴/벤치', '영아용의자', '스텝스툴/사다리', '어린이의자','패브릭암체어', '천연가죽암체어', '인조가죽암체어', '리클라이너', '라탄암체어', '어린이암체어', '가죽암체어', '패브릭카우치', '천연가죽카우치']
chiffonier = ['서랍장', '어린이서랍장', '선반유닛', '수납장', '장식장', '선반/스탠드', '수납가구','화장대','기저귀교환대']
curtain = ['커튼', '블라인드', '샤워커튼', '패널커튼']




def learning(search,answer):
  search_token = []
  various = ['VA','NNG','NNP','VV']
  category_full = {'소파':sofa,'침대':bed,'옷장':closet,'테이블':table,'의자':chair,'장':chiffonier,'커튼':curtain}
  category_full_n = [0 for _ in range(7)]
  category = ''
  count = 0
  #print(search)
  searched_result = []
  if search in furniture_category:
    #print("검색어와 유사한 가구 1 : " + search)
    searched_result.append(search)
  else:
    search_sentence = mecab.pos(search)
    for word in search_sentence:
        for _ in various:
          if _ in word[1]:
            if model.wv.vocab.get(word[0]) != None:
                search_token.append((word[0]))
    #print("검색어 : " + str(search_token) + "\n")
    try:   
        search_results = model.wv.most_similar(positive=search_token, topn=500)        
        n = 0
        for search_result in search_results:
            if search_result[0] in furniture_category:
                #print(search_result)
                n += 1
                temp_c = 0
                for _ in category_full.keys():
                  if search_result[0] in category_full[_]:
                    category_full_n[temp_c]+=1
                  temp_c +=1 
                if n == 5: break
        
        #print('\n')
        #category_full_n = sorted(category_full_n, reverse=True)
        # print(category-full_n)
        if category_full_n[0] == category_full_n[1]:
            for search_result in search_results:
                if search_result[0] in furniture_category:
                    for _ in category_full.keys():
                      if search_result[0] in category_full[_]:
                        category = _
                    break
            for search_result in search_results:
                if search_result[0] in category_full[category]:
                    count += 1
                    #print("검색어와 유사한 가구 " + str(count) + " : " + search_result[0])
                    searched_result.append(search_result[0])
                    if count == 3: break
        else:
            maxValue = max(category_full_n)
            temp_c = 0
            for u in category_full.keys():
              if maxValue == category_full_n[temp_c]:
                for search_result in search_results:
                    if search_result[0] in category_full[u]:
                        count += 1   
                        #print("검색어와 유사한 가구 " + str(count) + " : " + search_result[0]) 
                        searched_result.append(search_result[0])
                        if count == 3: break
              temp_c +=1                       
    except:
        #print("검색어를 다시 입력해주세요.")
        pass
  if answer in searched_result:
    return 1
  else:
    print(answer,searched_result)
    return 0
```
